```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Distribución normal teórica


$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

* $\mu$: media de la distribución
* $\sigma$: desviación estándar de la distribución

```python
def gaussian(x, mu, sigma):
  return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma,2))
```

```python
x = np.arange(-4, 4, 0.1);
y = gaussian(x, 0.0, 1.0);

plt.plot(x, y)
```

```python
x = np.arange(-4, 4, 0.1)
y = gaussian(x, 0.0, 1.5)

plt.plot(x, y)
```

```python
# distrbucion de densidad
dist = norm(0, 1)
x = np.arange(-4, 4, 0.1)
# pdf , probability density funcion , funcion de densidad de probabilidad
y = [dist.pdf(value) for value in x]

plt.plot(x, y)
```

```python
# distrbucion de densidad acumulada
dist = norm(0, 1)
x = np.arange(-4, 4, 0.1)

y = [dist.cdf(value) for value in x]

plt.plot(x, y)
```

## Distribución normal (gausiana) a partir de los datos

* *El archivo excel* lo puedes descargar en esta página: https://seattlecentral.edu/qelp/sets/057/057.html


```python
df = pd.read_excel('./values.xls')
df
```

```python
arr = df['Normally Distributed Housefly Wing Lengths'].values[3:]
arr
```

```python
values , dist = np.unique(arr, return_counts=True)
print(values)
print(dist)
plt.bar(values ,dist)
```

```python
# estimacion de distribucion

mu = arr.mean()

sigma = arr.std()

x = np.arange(30, 60, 0.1)

dist = norm(mu, sigma)

y = [dist.pdf(value) for value in x]

values, dist = np.unique(arr, return_counts=True)

plt.plot(x,y)
plt.bar(values, dist/len(arr))
```
